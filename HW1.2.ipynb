```python
import csv
from sklearn import linear_model
import numpy as np
import pandas as pd
```

```python
#Load Data
data =  pd.DataFrame.from_csv('/Users/leighajarett/Desktop/titanic_train.csv')
test_data = pd.DataFrame.from_csv('/Users/leighajarett/Desktop/titanic_test.csv')
```

```python
#data.Sex.replace(['female','male'],['0','1'],inplace=True) #make gender binary
#test_data.Sex.replace(['female','male'],['0','1'],inplace=True)

X_train = [data.Pclass, data.Sex, data.SibSp, data.Parch] #include passenger class and sex 
X_train = np.asarray(X_train) 
X_train = X_train.transpose()
Y_train = np.asarray(data.Survived)

X_test = [test_data.Pclass, test_data.Sex, test_data.SibSp, test_data.Parch]
X_test = np.asarray(X_test)
X_test = X_test.transpose()
```

```python
cls = linear_model.LogisticRegression()
fit = cls.fit(X_train, Y_train)
```

```python
fit.predict(X_test)
```
